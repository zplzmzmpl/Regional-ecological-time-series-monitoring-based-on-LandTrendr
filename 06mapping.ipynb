```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib_scalebar.scalebar import ScaleBar
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
# Open the image file
with rasterio.open('/content/51years_mul_cls_5.tif') as src:
    data = src.read(1)

img_extent = [-12492990, -12474390, 4936290, 4955130]
# Define the color map
colors = [(0, 0, 1), (0, 1, 0), (153/255, 0, 0), (255/255, 204/255, 204/255), (72/255, 72/255, 72/255)]
N = len(colors)

cmap_name = 'custom_map'
cm = mcolors.LinearSegmentedColormap.from_list(cmap_name, colors, N=N)

# Plotting
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'projection': ccrs.epsg('3857')})

# Plot the image
ax.imshow(data, origin='upper', extent=img_extent, cmap=cm, interpolation='nearest', transform=ccrs.epsg('3857'))

# Add legend
labels = ['Cluster1', 'Cluster2', 'Cluster3', 'Cluster4', 'Cluster5']

patches = [plt.plot([],[], marker="s", ms=10, ls="", mec=None, color=colors[i],
            label="{:s}".format(labels[i]))[0] for i in range(len(labels))]
plt.legend(handles=patches, bbox_to_anchor=(1.05, 1), title = 'categories', loc=2, borderaxespad=0.)

# Add scale bar
# Note: You need to replace `scale` with the appropriate scale in meters per pixel
scale = 30  # Placeholder value
ax.add_artist(ScaleBar(scale))

# Add graticules
ax.gridlines(draw_labels=True)

# Add compass
# Note: Compass may not be accurate on maps not oriented to true north
plt.annotate('KShape', xy=(0.1, 0.95), xycoords='axes fraction', ha='center',
             va='center', fontsize=20, rotation=0)

# Add other geographic features
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.set_title('Multivariates KShape Classification Map')

# Save the figure
plt.savefig('mulitivariates_kshape_5_map.png', bbox_inches='tight')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from matplotlib.patches import Patch
import rasterio
import numpy as np

# Open the image file
with rasterio.open('/content/slc-change-map-1972-2024.tif') as src:
    data = src.read(2)

img_extent = [-12492990, -12474390, 4936290, 4955130]

# Define the number of levels
num_levels = 10  # 

valid_values = data[np.isfinite(data)]

# Calculate the boundaries based on the min and max values in the data
boundaries = np.linspace(np.min(valid_values), np.max(valid_values), num_levels + 1)

# Generate a list of colors for each level
colors = [mcolors.to_hex(plt.cm.viridis(i / num_levels)) for i in range(num_levels)]

# Create a custom color map
cmap_name = 'custom_map'
cm = mcolors.LinearSegmentedColormap.from_list(cmap_name, colors, N=num_levels)

# Create a normalization instance for imshow
norm = mcolors.BoundaryNorm(boundaries, cm.N)

# Plotting
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'projection': ccrs.epsg('3857')})

# Plot the image
ax.imshow(data, origin='upper', extent=img_extent, cmap=cm, norm=norm, interpolation='nearest', transform=ccrs.epsg('3857'))

scale = 30  # Placeholder value
ax.add_artist(ScaleBar(scale))

# Add graticules
ax.gridlines(draw_labels=True)

# Add legend
patches = [Patch(color=colors[i], label=f'{boundaries[i]:.2f} - {boundaries[i+1]:.2f}') for i in range(num_levels)]
plt.legend(handles=patches, bbox_to_anchor=(1.05, 1), loc='upper left',title = 'Absolute Magnitude', borderaxespad=0.)

# Add other geographic features
# ax.add_feature(cfeature.COASTLINE)
# ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.set_title('Magnitude Map')

# Show the plot
plt.show()
```

# Folium

```python
import folium
import rasterio
from matplotlib import cm
from matplotlib.colors import to_hex

# GeoTIFF
with rasterio.open('/content/slc-change-map-1972-2024.tif') as src:
    band1 = src.read(1)

# Folium
m = folium.Map(location=[40.55, -112.13], zoom_start=10)

# 
unique_values = np.unique(band1)

# 
colormap = cm.get_cmap('viridis', len(unique_values))

# 
def scalar_to_color(scalar):
    norm_scalar = (scalar - unique_values.min()) / (unique_values.max() - unique_values.min())
    rgba = colormap(norm_scalar)
    return to_hex(rgba)

folium.raster_layers.ImageOverlay(
    image=band1,
    bounds=[[-12492990, -12474390], [4936290, 4955130]],
    colormap=scalar_to_color,
    opacity=0.8,
    interactive=True,
    cross_origin=False,
    zindex=1,
).add_to(m)

m
```

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import matplotlib.colors as cor
import numpy as np
import gma.extend.mapplottools as mpt
PAR = {'font.sans-serif': 'Times New Roman', 'axes.unicode_minus': False, }
plt.rcParams.update(PAR)
```

```python
InFiles = ["__ESA_2020.tif", "__ESA_2020 - .tif", "__ESA_2020 -  (2).tif", "__ESA_2020 -  (3).tif"]
#
CMap = []
Colors = []
for InFile in InFiles:
  DataSet = gma.Open(InFile)
Color = DataSet.GetGDALDataset().GetRasterBand(1).GetColorTable()
ColorTable = [Color.GetColorEntry(i) for i in range(Color.GetCount())]
#    Matplotlib 
CMapV = tuple(tuple(np.array(CT) / 255) for CT in ColorTable)
# 
CMap.append(cor.ListedColormap(CMapV))
Colors.append([CMapV[i] for i in range(10, 110, 10)] + [CMapV[95]])
##
Method = ['', '', '', '']
# 
ColorName = ['', '', '', '', '', '/', '', '', '', '', '']
```

```python
BlockSize = 8000
Columns = DataSet.Columns
Rows = DataSet.Rows
Blocks = [(r, c) for r in range(0, Rows, BlockSize) for c in range(0, Columns, BlockSize)]
```

```python
GEOT = DataSet.GeoTransform
Columns = DataSet.Columns
Rows = DataSet.Rows
# 
ExtentData = [GEOT[0], GEOT[0] + GEOT[1] * Columns, GEOT[3] + GEOT[-1] * Rows, GEOT[3]]
# 
EL, ER, EB, ET = 0.2, 0.1, 0.15, 0.05
# 
ExtentPLT = [ExtentData[0] - (ExtentData[1] - ExtentData[0]) * EL,
             ExtentData[1] + (ExtentData[1] - ExtentData[0]) * ER,
             ExtentData[2] - (ExtentData[3] - ExtentData[2]) * EB,
             ExtentData[3] + (ExtentData[3] - ExtentData[2]) * ET]
```
